```python
from sklearn.datasets import load_boston
X,y = load_boston(return_X_y =True)
```

```python
X[0]
```

```python
X.shape
```

```python
y.shape
```

```python
y
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
help(Sequential)
```

```python
# initilized the model
model = Sequential()
model.add(Dense(12,input_dim=X.shape[1],activation = "relu")) # input layer
model.add(Dense(12,activation="relu")) # Hidden Layer
model.add(Dense(1,activation="linear")) # output layer
```

```python
# compile the model
model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mean_absolute_percentage_error"]) # optimizwer
```

```python
model.fit(X,y, epochs= 2500, batch_size=100)
```

```python
accuracy = model.evaluate(X,y)
```

```python
model.summary()
```

```python
from sklearn.datasets import load_breast_cancer
X, y = load_breast_cancer(return_X_y=True)
```

```python
y
```

```python
X[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(15,input_dim =X.shape[1], activation = "relu"))
model.add(Dense(15, activation="relu"))
model.add(Dense(1, activation= "sigmoid"))
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(X,y, epochs= 150, batch_size=10)
```

```python
accuracy = model.evaluate(X,y)
```
